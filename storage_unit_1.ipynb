```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# prepare some coordinates
x, y, z = np.indices((10, 10, 10))

# height x width X length
# desk : 2.4 x 4.5 x 2'
# hutch : 7.25 x 5 x 2.16
```

```python
desk = (x < 2) & (y < 5) & (z < 3)
hutch = (x < 3) & (y >= 5) & (z < 8)
tote_1 = (x >= 7) & (y >= 8) & (z < 2)
tote_2 = (x >= 7) & (y < 2) & (z < 2)
tote_3 = (x >= 7) & (y >= 8) & (z >= 2) & (z < 4)
tote_4 = (x >= 7) & (y >= 8) & (z >= 4) & (z < 6)
tote_5 = (x >= 7) & (y >= 8) & (z >= 6) & (z < 8)
tote_6 = (x >= 4) & (x < 7) & (y >= 8) & (z >= 0) & (z < 2)
tote_7 = (x >= 4) & (x < 7) & (y >= 8) & (z >= 2) & (z < 4)
tote_8 = (x >= 4) & (x < 7) & (y >= 8) & (z >= 4) & (z < 6)
tote_9 = (x >= 4) & (x < 7) & (y >= 8) & (z >= 6) & (z < 8)

# combine the objects into a single boolean array
voxelarray = desk | hutch | tote_1 | tote_2 | tote_3 | tote_4 | tote_5 | tote_6 | tote_7 | tote_8 | tote_9

# set the colors of each object
colors = np.empty(voxelarray.shape, dtype=object)
colors[desk] = 'blue'
colors[hutch] = 'green'
colors[tote_1] = 'yellow'
colors[tote_2] = 'pink'
colors[tote_3] = 'black'
colors[tote_4] = 'yellow'
colors[tote_5] = 'black'
colors[tote_6] = 'black'
colors[tote_7] = 'yellow'
colors[tote_8] = 'black'
colors[tote_9] = 'yellow'

# and plot everything
ax = plt.figure().add_subplot(projection='3d')
ax.voxels(voxelarray, facecolors=colors, edgecolor='k')
# plt.rcParams["figure.figsize"] = (5,5)
# plt.rcParams["figure.figsize"] = plt.rcParamsDefault["figure.figsize"]
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)


plt.show()
```
